Below graph depicts how the connections were created throughout the time.

```python
lkd = lkd.sort_values(by = 'Connected On')

import plotly.express as px

px.line(lkd.groupby(by = 'Connected On').count().reset_index(),
        x = 'Connected On',
        y = 'First Name',
        labels = {'First Name': 'Number of Connections'},
        title = 'Connection Timeline'
        )
```

Grouping the data by Company and then sorting the results in descending order provides a brief overview of the places where most of my LinkedIn connections are working. Top places include, Middle East Technical University (My Undergraduate Institute), Ford Otsan, GUNSEL etc. 

```python
group_company = lkd.groupby(by = 'Company').count().reset_index()
group_company
```

```python
group_company = group_company.sort_values(by = 'Connected On', ascending = False).reset_index(drop = True)

group_company
```

A bar braph depicting number of employees in each company (top 50 only) is shown below.

```python
fig = px.bar(group_company[:50],
       x = 'Company',
       y = 'Connected On',
       labels = {'Connected On': 'Number of Connections'},
       width = 1000,
       height = 900,
       title = 'Bar Graph for Companies that my Connections are working at.'
      )
fig
```

A treemap to depict the same information is shown below. 

```python
fig2 = px.treemap(group_company[:50], path = ['Company', 'Position'],
       values = 'Connected On',
       labels = {'Connected On': 'Number of Connections'},
       width = 1000,
       height = 900,
       title = 'Tree Map for Companies that my Connections are working at.'
      )
fig2
```

Different Job Positions of my connections are shown below, for instance intern, software engineer. Actually, it will make more sense to ignore the bottom entries which a alot of 1s.
